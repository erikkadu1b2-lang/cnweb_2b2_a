equação de torricelli

```
import math

def caltorricelli(vzero, acel, des):
  t=2*acel*des
  v=vzero+t
  res = round(math.sqrt(v), 2)
  return res
```

teste de torricelli

```
print("calculo de torricelli:" ,caltorricelli(0,1,8))
```
