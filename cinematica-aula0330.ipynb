equação de torricelli

```
import math

def caltorricelli(vzero, acel, des):
  t=2*acel*des
  v=vzero+t
  res = round(math.sqrt(v), 2)
  return res
```

teste de torricelli

```
print("calculo de torricelli:" ,caltorricelli(0,1,8))
```

formulas do movimentos uniformemente variado (muv)

```
def calmovimentouniforme(vinicial, acel, tempo):
  vh = vinicial + (acel * tempo)
  return vh
```

```
print("velocidade horaria:", calmovimentouniforme(12. 15, 78), "ms")
```
